### HDFS 데이터 READ / WRITE

```python
import pandas as pd
from hdfs import InsecureClient
import os
```

```python
client_hdfs = InsecureClient('http://192.168.137.100:50070')
```

```python
selloutData = pd.read_csv("../dataset/kopo_product_volume.csv")
```

```python
selloutData.head(10)
```

### 데이터 저장하기

```python
# Writing Dataframe to hdfs
with client_hdfs.write('/kopo/hongde_seminar.csv', encoding = 'utf-8', overwrite=True) as writer:
    selloutData.to_csv(writer)
```

```python
# ====== Reading files ======
with client_hdfs.read('/kopo/hongde_seminar.csv', encoding = 'utf-8') as reader:
    readdf=pd.read_csv(reader,index_col=0)
```
