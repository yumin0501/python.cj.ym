```python
inputList=[10,100,200,1000,1,2]
```

```python
def listAverage(inputList):
    # inputList=[10,100,200,1000,1,2]

    minvalue = min(inputList)
    maxvalue = max(inputList)

    inputList.remove(minvalue)
    inputList.remove(maxvalue)

    average = sum(inputList) / len(inputList)
    
    return average
```
